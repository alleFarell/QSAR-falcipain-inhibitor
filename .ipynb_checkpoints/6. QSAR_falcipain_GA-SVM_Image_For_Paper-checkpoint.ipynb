## Hyperparameter Tuning Result Comparison Plot

```python
scores_linear_ = np.array([0.91809725, 0.29383928, 0.8403861 , 0.71258306, 0.23768219, 
                  0.6486529 , 0.56240086, 0.3885615 , 0.48432751, 0.62359804])
scores_linear_tuned = 0.6690184816138537

scores_rbf_ = np.array([0.88249408, 0.17481106, 0.7786089 , 0.65405646, 0.54116709, 
               0.52130018, 0.63207484, 0.85406868, 0.8887125 , 0.83277203])
scores_rbf_tuned = 0.7278016591371522

scores_poly_ = np.array([0.67116941, 0.28802848, 0.65382247, 0.3381288 , 0.6250243, 
                0.70022603, 0.20838029, 0.62494749, 0.82897715, 0.48848592])
scores_poly_tuned = 0.6906847082926204
```

```python
scores_linear_
```

```python
score_diff_linear = scores_linear_tuned - scores_linear_.mean()
score_diff_rbf = scores_rbf_tuned - scores_rbf_.mean()
score_diff_poly= scores_poly_tuned - scores_poly_.mean()
```

```python
HT_result = pd.DataFrame({'Program': ['Non Tuned', 'Tuned', 'Non Tuned', 'Tuned', 'Non Tuned', 'Tuned', 'Difference', 'Difference', 'Difference'],
                  'Model': ['Linear', 'Linear', 'RBF', 'RBF', 'Poly', 'Poly', 'Linear', 'RBF', 'Poly'],
                  'R2 Score': [scores_linear_.mean(),scores_linear_tuned,scores_rbf_.mean(),scores_rbf_tuned, 
                               scores_poly_.mean(),scores_poly_tuned, score_diff_linear, score_diff_rbf, score_diff_poly ]
                 })

HT_result['R2 Score'] = HT_result['R2 Score'].round(2)

plt.rcParams["figure.figsize"] = [6.00, 5]
plt.rcParams["figure.autolayout"] = True

ax = sns.barplot(data=HT_result, x='Model', y='R2 Score', hue='Program')


ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[2])

plt.legend(fontsize=8.8, bbox_to_anchor= (0.56,1))
plt.savefig('Analysis Plot Images/Images in Paper/HT_result_comparison_1000_dpi.png', dpi=1000)
plt.show()
```

## Genetic Algorithm Convergence Plot

```python
fig = solutions_linear_df.sort_values(by='objective_scores', ascending=False, ignore_index=True)['plot'][0]
dict_linear =fig.to_dict()

fig = solutions_rbf_df.sort_values(by='objective_scores', ascending=False, ignore_index=True)['plot'][0]
dict_rbf =fig.to_dict()

fig = solutions_poly_df.sort_values(by='objective_scores', ascending=False, ignore_index=True)['plot'][0]
dict_poly =fig.to_dict()


plt.rcParams["figure.figsize"] = [5.70, 4.50]
plt.rcParams["figure.autolayout"] = True
sns.lineplot(x=dict_linear['data'][1]['x'], y=dict_linear['data'][1]['y'], label='Linear')
sns.lineplot(x=dict_rbf['data'][1]['x'], y=dict_rbf['data'][1]['y'], label='RBF')
sns.lineplot(x=dict_poly['data'][1]['x'], y=dict_poly['data'][1]['y'], label='Poly')
plt.xlabel("GA Generation")
plt.ylabel("Objective Score")
plt.legend()
# plt.savefig('Analysis Plot Images/convergence_plot_GA.png', dpi=1000, bbox_inches="tight")
plt.show()
```
