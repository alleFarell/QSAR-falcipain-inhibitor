## Data Splitting and Export Preprocessed Dataset

```python
### Data Splitting

from sklearn.model_selection import train_test_split

X = df_falcipain_inhibitor.drop(['Name', 'FP-2 pIC50 (uM)'], axis=1)
y = df_falcipain_inhibitor['FP-2 pIC50 (uM)']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Feature Selection using Genetic Algorithm from ZOOFS Library

```python
from zoofs import GeneticOptimization
```

```python
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.metrics import r2_score
# define your own objective function, make sure the function receives four parameters,
#  fit your model and return the objective value ! 
def objective_function_topass(model,X_train, y_train, X_test, y_test):      
    model.fit(X_train,y_train)  
    r2 = r2_score(y_test,model.predict(X_test))
    return r2

# create object of algorithm
algo_object=GeneticOptimization(objective_function_topass,n_iteration=20,
                            population_size=20,selective_pressure=2,elitism=4,
                            mutation_rate=0.1,minimize=False)



svr_model = make_pipeline(StandardScaler(), SVR(C=1.0, epsilon=0.2)) 

# fit the algorithm
algo_object.fit(svr_model,X_train, y_train, X_test, y_test,verbose=True)

#plot your results
algo_object.plot_history()

# extract the best  feature set
algo_object.best_feature_list 
```
